```python
import os 
import pandas as pd
import cv2
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, HTML
```

```python
curr_dir = os.getcwd()
excavators = os.path.join("curr_dir", "excator")

for excavator in os.listdir(excavators):
    img_array = cv2.imread(os.path.join(excavators, excavator), cv2.IMREAD_GRAYSCALE)
    plt.imshow(img_array, cmap="gray")
    plt.show()
    break
```

```python
IMG_SIZE = 50

newray = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
plt.imshow(newray, cmap="gray")
plt.show()
```

```python
training_data = []

def insertCATastrophy():
    k = 0
    for i in os.listdir(excavators):
        img_array = cv2.imread(os.path.join(excavators, i), cv2.IMREAD_GRAYSCALE)
        new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
        training_data.append([new_array, k])

insertCATastrophy()

print(len(training_data))
```

```python
import random

random.shuffle(training_data)

for sample in training_data:
    print(sample[0])
    break
```

```python
x = []
y = []

for imgbit, model in training_data:
    x.append(imgbit)
    y.append(model)

x = np.array(x).reshape(-1, IMG_SIZE, IMG_SIZE, 1)
```
